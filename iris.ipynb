# Training the model 

```python
X=data.drop(columns=["species"]) #features
y=data["species"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report
```

```python
lr_model=LogisticRegression(max_iter=500)
lr_model.fit(X_train,y_train)
```

```python
lr_pred=lr_model.predict(X_test)
print(f"accuracy_score : {accuracy_score(y_test,lr_pred)}")
print(f"Classification_report : {classification_report(y_test,lr_pred)}")
```

```python
dt_model=DecisionTreeClassifier()
dt_model.fit(X_train,y_train)
```

```python
dt_pred=dt_model.predict(X_test)
print(f"accuracy_score : {accuracy_score(y_test,dt_pred)}")
print(f"Classification_report : {classification_report(y_test,dt_pred)}")
```

```python
svm_model=SVC()
svm_model.fit(X_train,y_train)
```

```python
svm_pred=svm_model.predict(X_test)
print(f"accuracy_score : {accuracy_score(y_test,svm_pred)}")
print(f"Classification_report : {classification_report(y_test,svm_pred)}")
```
